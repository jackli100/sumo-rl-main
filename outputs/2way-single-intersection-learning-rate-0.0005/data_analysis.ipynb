```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path_1 = 'dqn_conn0_ep5.csv'
file_path_10 = 'dqn_conn0_ep6.csv'

data_1 = pd.read_csv(file_path_1)
data_10 = pd.read_csv(file_path_10)

# 
steps_1 = data_1['step']
total_stopped_1 = data_1['system_total_stopped']

# 
steps_10 = data_10['step']
total_stopped_10 = data_10['system_total_stopped']

# 
plt.figure(figsize=(10, 6))
plt.plot(steps_1, total_stopped_1, label='Total Stopped (File 1)', color='blue')
plt.xlabel('Step')
plt.ylabel('System Total Stopped')
plt.title('System Total Stopped vs Step (File 1)')
plt.legend()
plt.grid(True)
plt.show()

# 
plt.figure(figsize=(10, 6))
plt.plot(steps_10, total_stopped_10, label='Total Stopped (File 10)', color='green')
plt.xlabel('Step')
plt.ylabel('System Total Stopped')
plt.title('System Total Stopped vs Step (File 10)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path = 'dqn_conn0_ep1.csv'
data_sample = pd.read_csv(file_path)

# 100
data_sample_100 = data_sample.head(100)

# step
steps_sample = data_sample_100['step']
variables = data_sample_100.drop(columns=['step'])

# stepstep
plt.figure(figsize=(12, 8))
for column in variables.columns:
    plt.plot(steps_sample, variables[column], label=column)

plt.xlabel('Step')
plt.ylabel('Values')
plt.title('Variables vs Step (First 100 rows)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
